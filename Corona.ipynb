# Ordenar a data para grafico de 30 dias y guardarla

```python
lit_x2 = list(['Alemania', 'Argentina', 'Bolivia', 'Brasil', 'Chile', 'EE. UU.', 'Ecuador', 'España', 'Francia', 
               'Japón', 'Italia', 'Venezuela', 'Colombia', 'Paraguay', 'Perú', 'México', 'Uruguay', 'Reino Unido', 'China',
              'Irán', 'Corea del Sur'])
lit_x2.sort()
```

```python
bel = pd.DataFrame()

for i in lit_x2:
    
    bel2 = pd.DataFrame({'Country':[i], 'Casos':[0], 'Dia': [0]})
    
    X = derretido.loc[derretido['Country'].isin([i])]
    X.sort_values(["Casos"], axis = 0, ascending = True, inplace = True)
    X = X[X['Casos'] > 0]
    X['Dia'] = lb_make.fit_transform(X['Fecha'])
    X.sort_values(["Dia"], axis = 0, ascending = True, inplace = True)
    X["Dia"] = X["Dia"].apply(lambda x: x+1)
    X.drop(['Fecha'], axis=1, inplace=True)
    X = X.iloc[0:30, :]
    X = pd.concat([bel2, X], axis = 0, ignore_index = True, join = 'outer')
    bel = pd.concat([bel, X], axis = 0, ignore_index = True, join = 'outer')
    bel.sort_values(['Country', 'Dia'], axis = 0, ascending = True, inplace = True)
    X = pd.DataFrame()
```

```python
bel.to_csv('data.csv', index = False, encoding = "UTF-8")
```

# Para el mapa

```python
## Aca se obtendran las cordenadas long lat de cada pais
def geolocate(country = None):
    loc = geolocator.geocode(country)
    return (loc)
```

```python
latitud = []
longitud = []

for i in lit_x2:
    
    loc = geolocator.geocode(i, timeout=10)
    latitud.append(loc.latitude)
    longitud.append(loc.longitude)
    
total_tabla['Latitud'] = latitud
total_tabla['Longitud'] = longitud
```

```python
total_tabla.to_csv('data_casos_totales2.csv', index = False, encoding = "UTF-8")
```

## Parte para Analisis de crecimiento

```python
# Para los casos enfermos

df_confirmed = pd.DataFrame()

for i in lit_x2:
    
    X = derretido.loc[derretido['Country'].isin([i])]
    X.sort_values(["Casos"], axis = 0, ascending = True, inplace = True)
    X.sort_values(["Fecha"], axis = 0, ascending = True, inplace = True)
    df_confirmed = pd.concat([df_confirmed, X], axis = 0, ignore_index = True, join = 'outer')
    df_confirmed.sort_values(['Country', 'Fecha'], axis = 0, ascending = True, inplace = True)
    X = pd.DataFrame()
```

```python
df_confirmed.shape
```

```python
# Para los personas fallecidos

df_dea = pd.DataFrame(data['deaths'])
df_dea.drop(['Province/State', 'iso_code', 'Lat', 'Long'], axis=1, inplace=True)
df_dea = df_dea.rename(columns={'Country/Region':'Country'})
df_dea.loc['R. Unido', "Country"] = 'Reino Unido'
df_dea = df_dea.groupby("Country").agg('sum').reset_index()
df_dea = df_dea.loc[df_casos['Country'].isin(lit_x2)]
df_dea = pd.melt(df_dea, id_vars = ['Country'], var_name ='Fecha', value_name ='Deaths')
```

```python
df_deaths = pd.DataFrame()

for i in lit_x2:
    
    X = df_dea.loc[df_dea['Country'].isin([i])]
    X.sort_values(["Deaths"], axis = 0, ascending = True, inplace = True)
    X.sort_values(["Fecha"], axis = 0, ascending = True, inplace = True)
    df_deaths = pd.concat([df_deaths, X], axis = 0, ignore_index = True, join = 'outer')
    df_deaths.sort_values(['Country', 'Fecha'], axis = 0, ascending = True, inplace = True)
    X = pd.DataFrame()
```

```python
df_deaths.to_csv('df_deaths.csv', index = False, encoding = "UTF-8")
```

```python
df_confirmed.shape, df_deaths.shape
```

```python
# para personas recuperadas

df_rec = pd.DataFrame(data['recovered'])
df_rec.drop(['Province/State', 'iso_code', 'Lat', 'Long'], axis=1, inplace=True)
df_rec = df_rec.rename(columns={'Country/Region':'Country'})
df_rec = df_rec.groupby("Country").agg('sum').reset_index()
df_rec = df_rec.loc[df_casos['Country'].isin(lit_x2)]
df_rec = pd.melt(df_rec, id_vars = ['Country'], var_name ='Fecha', value_name ='Recovered')
```

```python
df_recovered = pd.DataFrame()

for i in lit_x2:
    
    X = df_rec.loc[df_rec['Country'].isin([i])]
    X.sort_values(["Recovered"], axis = 0, ascending = True, inplace = True)
    X.sort_values(["Fecha"], axis = 0, ascending = True, inplace = True)
    df_recovered = pd.concat([df_recovered, X], axis = 0, ignore_index = True, join = 'outer')
    df_recovered.sort_values(['Country', 'Fecha'], axis = 0, ascending = True, inplace = True)
    X = pd.DataFrame()
```

```python
df_confirmed.shape, df_deaths.shape, df_recovered.shape
```

```python
## Unir los dataframes

dfaux = pd.merge(df_confirmed, df_deaths,  on=['Country', 'Fecha'])
df_totales = pd.merge(dfaux, df_recovered,  on=['Country', 'Fecha'])
df_totales = df_totales[['Country', 'Fecha', 'Casos', 'Deaths', 'Recovered']]
```

```python
df_totales.to_csv('df_totales.csv', index = False, encoding = "UTF-8")
```
